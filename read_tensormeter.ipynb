```python
from tensormeter import tensormeter
import time
import matplotlib.pyplot as plt

%matplotlib widget

def measure_data_point(tens, int_time=1):
    #tens.send_meas(-1)
    tens.send_cldt()
    t0 = time.time()
    print("integrate...")
    while time.time()-t0<int_time:
        pass
    print("request data")
    d,a = tens.get_all_data()
    print(d[:,9].mean())
    return d[:,9].mean()
```

```python
tens = tensormeter(0, HOST="134.30.1.138")
```

```python
datapoints = []
timestamps = []
iterators = []
i = 0
```

```python
d = measure_data_point(tens, int_time=3)
datapoints.append(d)
timestamps.append(time.time())
i += 1
iterators.append(i)
```

```python
fig,ax = plt.subplots(1,2)
ax[0].plot(iterators,datapoints)
ax[1].plot(timestamps,datapoints)
```
